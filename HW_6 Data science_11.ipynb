```python
import os
import pandas as pd

from google.colab import drive
drive.mount('/content/gdrive')

#Створюю дата сет для аналізу
path = '/content/gdrive/MyDrive/Colab Notebooks/data'

#read the data from the data_2d.csv into a dataframe
df_data_2d = pd.read_csv(path + '/' + 'data_2d.csv')

#read the data from the mnist.csv into a dataframe
df_data_mnist = pd.read_csv(path + '/' + '/mnist.csv')
```

```python
#1
#create an array from the df_data_2d dataframe without the labels with the to_numpy() method
X_data_2d = pd.DataFrame(df_data_2d).to_numpy()
print(f'X_data_2d: \n{X_data_2d}')

#create an array from the df_data_2d dataframe without the labels with the to_numpy() method
X_data_mnist = pd.DataFrame(df_data_mnist).to_numpy()
print(f'X_data_mnist: \n{X_data_mnist}')
```

```python
#2 create an empty list to store the values of the distortion/inertia
distortions_df_data_2d = []
distortions_df_data_mnist = []
```

```python
#3 create a for loop to iterate through the values of k from 1 to 10
from sklearn.cluster import KMeans

clusters = range(1, 10)
for k in clusters:
    #4 create an object of the KMeans class with the number of clusters
    kmeans = KMeans(n_clusters=k, random_state=42)
    #5 fit the model with the data
    kmeans.fit(X_data_2d)
    #6 append the value of the distortion/inertia to the list
    distortions_df_data_2d.append(kmeans.inertia_)
```

```python
#7 plot the values of the distortion/inertia
import matplotlib.pyplot as plt

plt.plot(clusters, distortions_df_data_2d, 'bx-')
plt.xlabel('k')
plt.ylabel('Inertia')
plt.title('The Elbow Method showing the optimal k')
plt.show()
```
